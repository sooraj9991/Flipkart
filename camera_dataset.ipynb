```python
df['Release date'].value_counts().sort_index().plot(kind='bar')   #plotting using bar graph
plt.show()
```

Number of models released based on years are increasing


```python
dataFrameCorr=df.corr(method='pearson')
plt.subplots(figsize=(8,8))                                                      # plotting correlation using heatmap from seaborn
sns.heatmap(dataFrameCorr, annot=True)
plt.title('Correlation between Attributes')
plt.show()
```

This shows the correlationship between the attributes


```python
GroupedByDate = df.groupby(['Release date'])               # grouping data according to the column Release date
```

```python
fig=plt.figure(figsize=(18,6))

ax=plt.subplot2grid((3,5),(0,0),colspan=4,rowspan=2)
GroupedByDate['Max resolution'].max().plot(kind="bar",color='green',legend=True,position=0,width=0.4)
GroupedByDate['Low resolution'].max().plot(kind="bar",color='purple',legend=True,position=1,width=0.4)                   # plotting using bar graph for both max and low resolution
plt.show()
```

We can see that resolutions are inhanced at every year


```python
fig=plt.figure(figsize=(18,6))

plt.subplot2grid((3,5),(0,0),colspan=4,rowspan=2)
GroupedByDate['Zoom wide (W)'].max().plot(kind="bar",color='green',legend=True,position=0,width=0.4)                      # plotting using bar graph for both the zooms(wide and tele)
GroupedByDate['Zoom tele (T)'].max().plot(kind="bar",color='red',legend=True,position=1,width=0.4)
plt.show()
```

the enhance in both type of Zoom (wide and tele) are increase in slow manner




```python
fig=plt.figure(figsize=(18,6))
GroupedByDate.max().pivot_table(index='Release date', columns='Price', values='Max resolution').plot.bar(stacked=True)           # plotting usimg bar graph
plt.ylabel("Max resolution")
plt.show()
```

This show relashipship between price and Max resolution on all years
